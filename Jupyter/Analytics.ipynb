## Load Modules

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

## Load Dataset

```python
df=pd.read_csv('Dataset/Preprocessed Dataset.csv')
df.head()
```

## Yearwise Car Data Analysis

```python
#Pie chart for year
year=df['Year'].value_counts()
year.plot(kind='pie',autopct='%1.1f%%',figsize=(20,20))
plt.legend()
plt.show()
```

## Average price of Car every year

```python
#average price of each year
year=df.groupby('Year')['Price'].mean()
year.plot(figsize=(10,10))
plt.ticklabel_format(style='plain', axis='y')
plt.legend()
plt.show()
```

## Owner number vs Price

```python
#Owner price relation
owner=df.groupby('Owner')['Price'].mean()
owner.plot(figsize=(10,10))
plt.legend()
plt.show()
```

## Seller Type vs Price

```python
#Seller type price relation with big size
seller=df.groupby('Seller_Type')['Price'].mean()
seller_type=['Corporate', 'Individual', 'Commercial Registration', 'Dealer',
       'Trustmark Dealer']
plt.bar(seller_type,seller)
plt.ticklabel_format(style='plain', axis='y')
plt.xticks(rotation=90)
plt.xlabel('Seller Type')
plt.ylabel('Price')
plt.show()
```

## Boxplot

```python
# distance price relation
distance=df.groupby('Distance')['Price'].mean()
distance.plot(figsize=(10,10))
plt.ticklabel_format(style='plain', axis='y')
plt.legend()
plt.show()
```
